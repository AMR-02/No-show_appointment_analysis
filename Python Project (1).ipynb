```python
# Checking the distribution of Age using boxplot:
app_df.Age.plot(kind="box")
# Shows that most of the ages fall under the range of 19-58
```

```python
# Defining a function to create new column called 'Age_category':
def Age_cat(age):
    if 0 < age <= 14:
        return "0-14"
    elif 14 < age <= 24:
        return "15-24"
    elif 24 < age <= 40:
        return "25-40"
    elif 41 < age <= 64:
        return "41-64"
    elif age >= 65:
        return "65 & older"
```

```python
app_df["Age_category"] = app_df.Age.apply(Age_cat).astype("category")
```

```python
app_df
```

```python
app_df.info()
```

```python
# Creating a variable to include columns that need to be grouped
cols_to_grp = ["Age_category", "Gender", "No_show"] 
```

```python
# Filtering the rows with "Missed Appointments"
missed_app = app_df[app_df.No_show == "Missed Appointment"] 
#Single mention of app_df creates a boolean expression. Not what we want
# What we want is; the rows from app_df where value in app_df.No_show is "Missed Appointment".
```

```python
#Checking the count of Missed Appointment by Age_category and Gender
missed_counts = missed_app[cols_to_grp].groupby(["Age_category", "Gender"],observed=True).value_counts()
missed_counts.to_frame()
```

```python
# Total count of appointments (missed and non-missed)
total_app = app_df.No_show.value_counts()
total_app
```

```python
total_app.sum()
```

```python
# Total Missed Appointments
total_missed = missed_counts.sum()
total_missed
```

```python
# Percentage of Missed Appointments out of total appointments
per_total_missed = (21372/110527)*100
per_total_missed
# Out of all the appointments, 20% of the appointments were missed. 
```

```python
# Percentage of Missed Appointments by Age and Gender
percent_missed = (missed_counts/total_missed*100).to_frame().sort_values(by='count', ascending=True)
percent_missed
```

```python
# Want to edit the values on y-axis:
y_values = ("41-64, F", "25-40, F", "15-24, F", "0-14, M", "0-14, F","41-64, M", "65 & older, F", "25-40, M", "15-24, M", "65 & older, M")
```

```python
# Creating plot (bar) for visualization
percent_missed.plot(kind='barh', title='Percentage of Missed Appointments by Age group and Gender', xlabel='Percentage of Missed Appointments', ylabel='Age group & Gender')
```

### **Conclusion:**
- The analysis shows that around 20% of all the appointments have been missed by patients.
- Missed appointments across gender and age groups were analyzed and following key observations were made:
  1. Women rank the highest in terms of missing medical appointments across all age groups, with majority of them falling in the category of age 41-64 and 25-40 (25-64).
  2. The higher no-show rates among women could possibly be due to factors such as work, care-giving responsibilities or a sensitive health condition.
  3. Men have considerably lower rates of missed appointments, with the least rate being men aged 65 years and older.

### **Recommendation:**
- Age & gender are important factors when it comes to appointment attendance. Healthcare providers could investigate the cause of women missing aged 25-65 since they have the higher no-show rates.
- Flexible appointment schedules could be considered for working women.
